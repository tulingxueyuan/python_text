## 字符串的简单操作

 +  字符串的连接操作
 *  字符串的复制操作
 [] 字符串的索引操作
 [: :]  字符串的切片操作
 [开始索引:结束索引:间隔值]  (包含开始,不包含结束)

```python
str1 = 'lo'
str2 = 've'
print(str1+str2)
print('lo'+'ve')

print('#'*10)

s = 'I like dog'
print(s[-1])
s = '012345678'
# 指定范围切字符串
print(s[2:5])
# 指定范围并且给出间隔值
print(s[2:6])
print(s[2:6:2])
# 切整个字符串
print(s[:])
# 指定开始,不指定结束
print(s[2:])
# 指定结束,不指定开始
print(s[:4])
# 指定开始,不指定结束并给出间隔值
print(s[2::3])
# 指定结束,不指定开始并给出间隔值
print(s[:6:2])
```

```python
# capitalize()  首字母大写   返回的是字符串
s = '狗 like dog'
print(id(s))
#new_s = s
new_s = s.capitalize()
print(new_s)
print(id(new_s))
```

```python
# title()  将每个单词的首字母变为大写   返回的是字符串
s = 'I 狗like dog'
print(s.title())
```

```python
# upper()  将所有字母变为大写字母     返回的是字符串
s = 'I 狗like dog'
print(s.upper())
```

```python
# lower()  将所有字母变为小写字母    返回的也是字符串
s = 'I 狗like dog'
print(s.lower())
```

```python
# swapcase()  大小写互换   返回的是字符串
s = 'I 狗like dog'
print(s.swapcase())
```

```python
# len() 计算字符串长度,不属于字符串的内建函数
s = 'I 狗like dog'
s1 = 'I like dog'
print(len(s))
print(len(s1))
```

```python
# find()  查找指定字符串,找不到返回-1  第一次找到返回第一次索引值
# index() 查找指定字符串,找不到报错
s = 'asdfghjklasdfghjkl'
s1 = s.find('s',3)
s2 = s.index('a')
print(s1)
print(s2)
```

```python
# count()  计算字符串出现次数  返回整形
s = 'asdfghjklasdfghjklsssss'
print(s.count('s'))
```

```python
# startswith() 检测是否以指定字母开头   返回布尔值
# endswith()  检测是否 以指定字母结束
s = 'I like dog'
print(s.startswith('i'))
print(s.startswith('I'))
print('='*20)
print(s.endswith('g'))
print(s.endswith('o'))
```

```python
# isupper()  检测所有字母是否是大写字母   返回的都是布尔值
s = 'f狗gh'
s1 = 'DF狗GH'
print(s.isupper())
print(s1.isupper())
print('='*20)
# islower()  检测所有字母是否是小写字母
print(s.islower())
print(s1.islower())
print('='*20)
# istitle()   检测是否以指定标题显示(每个单词首字母大写)
print(s.istitle())
print(s1.istitle())
s2 = 'I Like Dog'
print(s2.istitle())
print('='*20)
# isspace()  检测字符串是否是空字符串
s = '    '
s1 = 'i like   '
s2 = ' ' #至少有一个,否则返回False
print(s.isspace())
print(s1.isspace())
print(s2.isspace())
```

```python
# isalpha() 检测字符串是否是字母组成   返回布尔值
s = 'I 狗like dog'
s1 = 'I狗likedog'
print(s.isalpha())
print(s1.isalpha())
```

```python
# isalnum() 检测字符串是否有字母加数字组成    返回布尔值
s = 'I 狗li414ke dog'
s1 = 'I狗likedog132132'
s2 = '3133'
print(s.isalnum())
print(s1.isalnum())
print(s2.isalnum())
```

```python
# isdigit()  检测字符串是否由数字组成
# isdecimal()
# isnumeric()
s = '123'
print(s.isdigit())
print(s.isdecimal())
print(s.isnumeric())
print('='*20)
s = b'101100'
print(s.isdigit())
#print(s.isdecimal())
#print(s.isnumeric())
print('='*20)
s = '123.2'
print(s.isdigit())
print(s.isdecimal())
print(s.isnumeric())
print('='*20)
s = '三壹佰'
print(s.isdigit())
print(s.isdecimal())
print(s.isnumeric())
print('-'*20)
s = 'Ⅲ'
print(s.isdigit())
print(s.isdecimal())
print(s.isnumeric())
```

```python
# split() 用指定字符切割字符串   返回由字符串组成的列表
s = '日照香炉生紫烟*疑是银河落九天*飞流直下三千尺'
list1 = s.split('*')
print(list1)
```

```python
# splitlines()  以换行切割字符串
s = '''日照香炉生紫烟\n疑是银\n河落九天
飞流直下三千尺'''
print(s.splitlines())
```

```python
# join() 将列表按照指定字符串连接    返回的是字符串
list1 = ['日照香炉生紫烟', '疑是银河落九天', '飞流直下三千尺']
s = '*'.join(list1)
print(s)
```

```python
# ljust()  指定字符串的长度,内容靠左边,不足的位置用指定字符填充,默认空格, 返回字符串
s = 'abc'
print(len(s))
print(s.ljust(5)+'aa')

# center()  指定字符串长度,内容居中,不足的位置用指定字符填充,默认空格, 返回字符串
print(s.center(5,'#'))

# rjust() 指定字符串的长度,内容靠右边,不足的位置用指定字符填充,默认空格, 返回字符串
print(s.rjust(5,'#'))
```

```python
# strip() 去掉左右两边指定字符,默认是去掉空格
# lstrip()  去掉左侧指定字符,默认空格
# rstrip()  去掉右侧指定字符,默认空格

s = '   abc '
print('---'+s.strip()+'--')
print('---'+s+'--')

s = 'aaabcc'
print(s.lstrip('a'))
print(s.lstrip('b'))

print(s.rstrip('b'))
print(s.rstrip('c'))
```

```python
# zfill() 指定字符串长度,内容靠右,不足的位置用0填充
s = 'abc'
print(s.zfill(5))
```

```python
# maketrans()  生成用于字符串替换的映射表
# translate()  进行字符串替换
s = '今天晚上我吃的是小炒肉,可好吃了'
table = s.maketrans('小炒肉','大白菜')
print(table)
print(s.translate(table))
```

```python
# pop()  根据索引移除列表内一个元素,不给索引默认移除最后一个  返回移除的那个值
list1 = [1,2,3,4,5]
#list1.pop(2)
#print(list1)
print(list1.pop())
print(list1)
print(list1.pop())
print(list1)
list1.pop()
print(list1)
list1.pop()
print(list1)
list1.pop()
print(list1)
list1.pop()
```

```python
# reverse()  列表反转
list1 = [1,2,3,4]
list1.reverse()
print(list1)
```

```python
# sort()  排序  默认从小到大
list1 = [5,2,4,6,1,9]
list1.sort()
print(list1)

# 从大到小
list1.sort(reverse=True)
print(list1)
```

```python
# clear() 清楚整个字典  返回None
dict1 = {'a':1,'b':2,'c':3}
dict1.clear()
print(dict1)
```

```python
# copy()  复制字典
dict1 = {'a':1,'b':2,'c':3}
dict2 = dict1.copy()
print(dict2)
```

```python
# fromkeys() 按照指定的序列为键创建字典,值都是一样的
list1 = ['a','b','c']
dict1 = {}.fromkeys(list1)
dict2 = {}.fromkeys(list1,3)
print(dict1,dict2)
```

```python
# get() 根据键获取指定的值   找不到的键如果设默认值则返回默认值,如果没设默认值,则返回None
dict1 = {'a':1,'b':2,'c':3}
print(dict1.get('b'))
print(dict1.get('d'))
print(dict1.get('d',4))
```

```python
# items() 将字典变成类似于元组的形式方便遍历
dict1 = {'a':1,'b':2,'c':3}
for k,v in dict1.items():
    print(k,v)
    
print(dict1.items())
```

```python
# keys()  遍历字典中中键
# values()  遍历字典中的值
```

```python
# pop()  移除字典中指定元素   返回键所对应的值,如果键不存在,则返回默认值,如果键找不到,没设默认值,就会报错
dict1 = {'a':1,'b':2,'c':3}
print(dict1.pop('a'))
#print(dict1)
print(dict1.pop('d',4))
print(dict1.pop('d'))
```

```python
# popitem()  移除字典的键值对  返回移除的键和值
dict1 = {'d':4,'a':1,'b':2,'c':3}
print(dict1.popitem())
```

```python
# setdefault()  在字典里添加一个元素
dict1 = {'d':4,'a':1,'b':2,'c':3}
print(dict1.setdefault('e',5))
print(dict1)
```

```python
# update() 修改字典中的值
dict1 = {'d':4,'a':1,'b':2,'c':3}
dict1.update({'a':3,'b':4,'e':6})
print(dict1)
```

```python
# pop()  随机弹出一个元素
a = {'a','b','f',4}
a.pop()
print(a)
```

```python
# remove()  删除集合中的某个值,如果这个值不在集合中会报错
a = {'a','b','f',4}
a.remove(4)
print(a)
a.remove(4)
```

```python
# discard()  删除集合中的某个值,如果这个值不在集合中什么也不做
a = {'a','b','f',4}
a.discard(4)
print(a)
a.discard(4)
print(a)
```

```python
# difference()  差集  
# difference_update()  区别就是第一个返回一个新的集合,第二个是把原来集合覆盖
set1 = {1,2,3,4,7}
set2 = {2,4,8,111,24}
set3 = set1.difference(set2)
print(set3)
print(set1)
```

```python
set1 = {1,2,3,4,7}
set2 = {2,4,8,111,24}
set3 = set1.difference_update(set2)
print(set3)
print(set1)
```
